```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Import Excel
df = pd.read_excel (r'Data for Q2 Comparison of means.xlsx')
df
df.dtypes
```

```python
df
```

```python
##Casts Avg cost to numeric (float)
df['Aug cost'] = pd.to_numeric(df['Aug cost'], errors='coerce')
df.dtypes
```

```python
##Calculatess average per column 
df.loc['AVG'] = df.mean()
df
```

```python
###Calculate the cost dif for July, Aug, Sept (Cost Dif = Cost - Expected Cost)
##Syntax adds new columns in current df for each nwe calculated cost dif 

df['July Cost Dif'] = df['July cost'] - df['July Expected cost']
df['Aug Cost Dif'] = df['Aug cost'] - df['Aug expected cost']
df['Sept Cost Dif'] = df['Sept cost'] - df['Sept expected cost']
df
```

```python
##Calculate the degree of freedom for each month

dof_july = int(df['July Cost Dif'].count()) - 1
dof_Aug = int(df['Aug Cost Dif'].count()) - 1
dof_Sept = int(df['Sept Cost Dif'].count()) - 1


dof_july
dof_Aug 
dof_Sept
```

```python
##Calculate AVG cost for each month

avgcost_July = float(df.loc[['AVG'],['July cost']].values)
avgcost_Aug = float(df.loc[['AVG'],['Aug cost']].values)
avgcost_Sept = float(df.loc[['AVG'],['Sept cost']].values)

avgcost_july
avgcost_Aug 
avgcost_Sept
```

```python
##Calculate average expected cost

exp_cost_july = float(df.loc[['AVG'],['July Expected cost']].values)
exp_cost_Aug = float(df.loc[['AVG'],['Aug expected cost']].values)
exp_cost_Sept = float(df.loc[['AVG'],['Sept expected cost']].values)

exp_cost_july
exp_cost_Aug
exp_cost_Sept 
```

```python
##Generate view range for calculating stdv
rows = range(0,10)
columns = ['July Cost Dif', 'Aug Cost Dif', 'Sept Cost Dif']

df1 = df.loc[rows, columns]
df1
```

```python
### Calculate Stdv for all cost values 
stdv_all_cost = df1.stack().std()
stdv_all_cost
```

```python
##Create new dataframe using rpevious variables, use data variables to turn variables into columns

data = [['July', stdv_all_cost, dof_july, avgcost_July, exp_cost_july], 
        ['August', stdv_all_cost, dof_Aug, avgcost_Aug, exp_cost_Aug],
       ['September', stdv_all_cost, dof_Sept, avgcost_Sept, exp_cost_Sept]]
      
df2 = pd.DataFrame(data, columns = ['Month', 'stdv_all_cost', 'N', 'AVG cost', 'AVG expected cost'])

df2

```

```python
## Calculate stdv of all cost/ root of num of patients

df2['sqrt(N)'] = np.sqrt(df2['N'])
df2
```

```python

df2['StDv'] = df2['stdv_all_cost']/ df2['sqrt(N)']

df2
```

```python
### Find TDistribution Val

tvalues = [3.169, 3.355, 3.250]
df2['T-value'] = tvalues
df2
```

```python
df2['Upper control limit'] = df2['AVG expected cost'] + (df2['T-value'] * df2['StDv'])
df2['Lower control limit'] = df2['AVG expected cost'] - (df2['T-value'] * df2['StDv'])
df2
```

```python
plt.plot('Month', 'AVG cost', data=df2, marker ='o', color ='orange')
plt.plot('Month', 'Upper control limit', data=df2, markersize = 0, color = 'red')
plt.plot('Month', 'Lower control limit', data=df2, markersize = 0, color = 'red')
plt.legend()
plt.title('Actual Cost by Month')
plt.xlabel('Month')
plt.ylabel('Cost (USD)')
```
