```python
###Import Libraries 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline



```

```python
##Import Excel Data and print file 

df = pd.read_excel (r'Data_Q2_Introduction.xlsx')
print (df)
```

```python
df.plot(x='Time', y='Observed value', kind = 'line',marker='o')
plt.xlabel('Time')
plt.ylabel('Observed value')
plt.hlines(y = 52.5, xmin = 1, xmax = 7, color = 'r',label='pre-intervention')
plt.hlines(y = 12.5, xmin = 1, xmax = 7, color = 'r')
plt.hlines(y = 52.5, xmin = 9, xmax = 18, color = 'r',linestyles='dashed',label='post-intervention')
plt.hlines(y = 12.5, xmin = 9, xmax = 18, color = 'r',linestyles='dashed')
plt.xticks(df['Time'], df['Time'])
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```
